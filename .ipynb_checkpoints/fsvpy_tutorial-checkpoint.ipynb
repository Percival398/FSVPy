# Locate streaks
Now, we can locate streaks in the image.  This accomplished using the function locate_streaks. This function identifies streaks in your image using a contour finding algorithm and then returns a list where each entry contains the coordinates of a streak.  locate_streaks first pre-processes the image with a bandpass filter and a gaussian blur (the streak properties are measured on the raw data).  It has a default value of 5 times the standard deviation in intensity as a contour intensity level, but this can be adjusted using the optional contour_value argument, see the API for details.  

```python
s = fsv.locate_streaks(image)   
```

To see the streaks we have found, we can use the function plot_streaks

```python
fsv.plot_streaks(s, image)
```

We now have a pandas dataframe which lists the properties of all of our identified streaks, which we can then use to filter out undesired streak identifications.  The inputs to filter_streaks are the streak properties dataframe and the image size, for the example data it is (2048,2048).  There are optional arguments to allow for turning off edge filtering, and to adjust the filtering parameters. 

We will filter all streaks that are less than 7 pixels in height, and whose aspect ratio is less than 2.  Additionally, we will filter out streaks which are within 150 pixels of the edge of our image, as they are likely to be incomplete.

```python
filtered_streaks24 = strk.filter_streaks(streak_properties24, (2048,2048),min_height = 7, 
                                         min_aspect = 2, edge_filter = 150)
```

```python
print(filtered_streaks24)
```

As we can see above, many of our streaks did not survive the filtering.  This is OK - we use quite aggressive filtering with the philosphy that it is wasy to take a lot of (noisy) data.  We can see which streaks survived by using the plot_streaks function.  It takes as arguments your image, the filtered streak properties dataframe, and the list of found streaks.

```python
strk.plot_streaks(image24, filtered_streaks24, streaks24)
```

As you can see above, the filtering largely elminated the out-of-focus streaks, as well as those too close to the edge.  Depending on your images, more or less filtering may be required.

# Refine streak properties: fit streaks
Now that we are happy with our identification and filtering, we are ready to more precisely extract our streak properties.  To do so, we will use the fit_streaks function.  This function uses the location of the streak to extract a piece of the original image, and then uses fitting function to extract streak characteristics.  There are optional arguments to either plot the fits as you go (set plot_fits to True) or to save the fit plots to a file (set filepath to a string giving the filepath rather than the default of None).  If the fitting does not converge, streak height and width are set to zero.  The averaging and other parameters for fitting can be set with optional arguments, see the API for details.

```python
fitted_streaks = strk.fit_streaks(image24, filtered_streaks24, plot_fits = True)
```
